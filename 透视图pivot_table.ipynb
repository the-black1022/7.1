```python
#pivot_table有四个最重要的参数index、values、columns、aggfunc
```

Index就是层次字段，要通过透视表获取什么信息就按照相应的顺序设置字段

```python
#将对手设置为index
pd.pivot_table(df,index=['对手'])
```

```python
#将胜负设置为index
pd.pivot_table(df,index=['胜负'])
```

```python
#将主客场设置为index
pd.pivot_table(df,index=['主客场'])
```

```python
#对手与主客场都设置为index，其实就变成为了两层索引
pd.pivot_table(df,index=[u'对手',u'主客场'])
```

我们只需要james harden在主客场和不同胜负情况下的得分、篮板与助攻三项数据

```python
pd.pivot_table(df,index=[u'主客场',u'胜负'],values=[u'得分',u'助攻',u'篮板'])
```

aggfunc默认为平均数，可以设置我们对数据聚合时进行的函数操作

```python
pd.pivot_table(df,index=[u'主客场',u'胜负'],values=[u'得分',u'助攻',u'篮板'],aggfunc=[np.sum,np.mean])
```

Columns类似Index可以设置列层次字段，它不是一个必要参数，作为一种分割数据的可选方式。
fill_value填充空值，默认为NAN,margins=True进行汇总，汇总的字段为‘All’

```python
pd.pivot_table(df,index=[u'主客场'],columns=[u'对手'],values=[u'得分'],aggfunc=[np.sum])
```

```python
pd.pivot_table(df,index=[u'主客场'],columns=[u'对手'],values=[u'得分'],aggfunc=[np.sum],fill_value=0)
```

```python
pd.pivot_table(df,index=[u'主客场'],columns=[u'对手'],values=[u'得分'],aggfunc=[np.sum],fill_value=0,margins=1)
```
